```python
def dapatBerita(link):
    import csv
    import urllib
    from urllib.request import urlopen
    from PIL import Image
    from io import StringIO
    import string
    import requests
    from bs4 import BeautifulSoup
    import time #memberi jeda antar crawling satu ke berikutnya
    
    html = urlopen(link)
    soup = BeautifulSoup(html, "lxml")
    
    judul = soup.find('h1', 'post_heading')
    
    news=[]
    beritas=soup.find_all('p')
    for berita in beritas:
        a=berita.text.strip()
        news.append(a)
    
    waktu= soup.find('div', 'read__time').text.strip()
    tanggal=waktu[12:23]
    
    gambars = soup.find_all("img")
    
    for gambar in gambars:
        if '/750x500/' in gambar['src']:
            r= gambar["src"]
            
    nama_file =  judul + ".jpg"
    urllib.request.urlretrieve(r, nama_file)
    
    ini_berita= [judul,news]
    time.sleep(10)
      
    return ini_berita
    
```
